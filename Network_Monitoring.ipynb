```python
import psutil
```

```python
net_if_addrs = psutil.net_if_addrs()
for interface_name, interface_addresses in net_if_addrs.items():
    for address in interface_addresses:
          print(f" address family: {address.family}")
          print(f" broadcast: {address.broadcast}")
          print(f" netmask: {address.netmask}")
          print(f" address: {address.address}")
          print()
```

```python
import psutil
net_if_stats = psutil.net_if_stats
for interface, stats in net_if_stats.items():
    print(f"interface: {interface}")
    print(f" isup: {stats.isup}")
    print(f" duplex: {stats.duplex}")
    print(f" speed: {stats.speed}MB")
    print(f" mtu: {stats.mtu}")
    print()
```

```python
# Get network connections
net_conns = psutil.net_connections()
for conn in net_conns:
    print(f"FD: {conn.fd}")
    print(f"Family: {conn.family}")
    print(f"Type: {conn.type}")
    print(f"Laddr: {conn.laddr}")
    print(f"Raddr: {conn.raddr}")
    print(f"Status: {conn.status}")
    print(f"PID: {conn.pid}")
    print()
```

```python
# Get network I/O counters
net_io_counters = psutil.net_io_counters()
print(f"Bytes Sent: {net_io_counters.bytes_sent}")
print(f"Bytes Received: {net_io_counters.bytes_recv}")
```
